Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

Answer:

Probabilistic modeling enables generative models to learn the probability distribution of the training data. Once this distribution is learned, the model can generate new samples that resemble the original data.

Difference between Generative and Discriminative Models
Generative Models	Discriminative Models
Learn the data distribution P(X) or joint distribution P(X,Y).	Learn the conditional probability P(Y|X).
Can generate new data.	Only classify or predict labels.
Examples: VAE, GAN, GPT	Examples: Logistic Regression, SVM, Decision Trees
Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

Answer:

Autoencoder	Variational Autoencoder (VAE)
Compresses data into a latent representation.	Learns a probability distribution over the latent space.
Used mainly for reconstruction.	Used for reconstruction and generation of new data.
Latent vector is deterministic.	Latent vector is probabilistic (mean and variance).
Cannot easily generate new samples.	Can generate new sentences by sampling latent variables.

VAEs are preferred for text generation because they produce diverse and meaningful outputs.

Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

Answer:

The attention mechanism allows the decoder to focus on the most relevant words in the input sentence while generating each output word.

Working:
The encoder converts the input sentence into hidden states.
During decoding, attention computes scores between the current decoder state and each encoder hidden state.
Higher scores indicate more important words.
A weighted combination of encoder outputs (context vector) is passed to the decoder.
Importance:
Handles long sentences effectively.
Improves translation accuracy.
Prevents loss of information.
Forms the basis of Transformer models.

Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction.

```python
texts = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Reshape
from tensorflow.keras.models import Model

# Dataset
texts = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
max_len = max(len(seq) for seq in sequences)

X = pad_sequences(sequences, maxlen=max_len, padding='post')

vocab_size = len(tokenizer.word_index) + 1

# Encoder
inputs = Input(shape=(max_len,))
x = Embedding(vocab_size, 8)(inputs)
x = Flatten()(x)
latent = Dense(16, activation='relu')(x)

# Decoder
x = Dense(max_len * 8, activation='relu')(latent)
x = Reshape((max_len, 8))(x)
outputs = Dense(vocab_size, activation='softmax')(x)

model = Model(inputs, outputs)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy')

model.summary()

Y = X.reshape(X.shape[0], X.shape[1], 1)

model.fit(X, Y, epochs=100, verbose=0)

pred = model.predict(X)

print("Training Complete")
```

Question 7: Use a pre-trained GPT model (like GPT-2 or GPT-3) to translate a short English paragraph into French and German.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

text = "Artificial Intelligence is transforming the world."

# French
input_text_fr = "translate English to French: " + text
inputs_fr = tokenizer(input_text_fr, return_tensors="pt")
outputs_fr = model.generate(**inputs_fr, max_new_tokens=50)
fr_translation = tokenizer.decode(outputs_fr[0], skip_special_tokens=True)

# German
input_text_de = "translate English to German: " + text
inputs_de = tokenizer(input_text_de, return_tensors="pt")
outputs_de = model.generate(**inputs_de, max_new_tokens=50)
de_translation = tokenizer.decode(outputs_de[0], skip_special_tokens=True)

print("Original:", text)
print("French:", fr_translation)
print("German:", de_translation)
```

Question 8: Implement a simple attention-based encoder-decoder model for English-to-Spanish translation using TensorFlow.

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, LSTM, Dense, Attention

# Encoder
encoder_inputs = tf.keras.Input(shape=(None,))
encoder_embed = Embedding(1000, 64)(encoder_inputs)

encoder_outputs, state_h, state_c = LSTM(
    128,
    return_sequences=True,
    return_state=True
)(encoder_embed)

# Decoder
decoder_inputs = tf.keras.Input(shape=(None,))
decoder_embed = Embedding(1000, 64)(decoder_inputs)

decoder_outputs = LSTM(
    128,
    return_sequences=True
)(decoder_embed,
  initial_state=[state_h, state_c])

# Attention
attention = Attention()([decoder_outputs, encoder_outputs])

outputs = Dense(1000, activation='softmax')(attention)

model = tf.keras.Model(
    [encoder_inputs, decoder_inputs],
    outputs
)

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy'
)

model.summary()
```

Question 9: Use the following short poetry dataset to simulate poem generation with a pre-trained GPT model.

```python
poems = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)

prompt = """
Roses are red, violets are blue,
Sugar is sweet, and so are you.
"""

result = generator(
    prompt,
    max_length=60,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```
